## EAZY-CDFS Run

Operation Files:

- eazy_zf-cdfs.ipynb (photometric redshift run)
- hd5_sedplot.ipynb (SED plot, which required hd5 generated in above ipynb)

Input Files:

- cdfs.v1.6.11.cat (primary catalogue file)
- cdfs.v1.6.1.zphot.param (parameter space for EAZY run)
- cdfs.v1.6.11.translate (translate catalogue bands to FILTER.RES.latest / FILTER.RES.latest.info)
- FILTER.RES.latest (filter response curves)

Output Files:

- cdfs.v1.6.1_zp_###.png (zero point analysis)
- cdfs.v1.6.1.data.fits (chi^2 values)
- cdfs.v1.6.1.zout.fits (redshift values)
- cdfs.v1.6.1.h5 (raw data for SED plots)
- cdfs.v1.6.1.zphot.zeropoint (zeropoints)
- cdfs.v1.6.1.zphot.translate (replica of input translate)

```python
%matplotlib inline
import os
import numpy as np
import importlib
import sys
import time
import eazy

import warnings
from astropy.utils.exceptions import AstropyWarning

np.seterr(all='ignore')
warnings.simplefilter('ignore', category=AstropyWarning)

print('EAZYCODE = '+ str(os.getenv('EAZYCODE')) + '\n')

print(time.ctime() + '\n')

print(sys.version + '\n')

for module in ['numpy', 'scipy', 'matplotlib','astropy','eazy', 'prospect']:
    mod = importlib.import_module(module)
    print('{0:>20} : {1}'.format(module, mod.__version__))
```

```python
os.getcwd()
translate_file = 'cdfs.v1.6.11.translate'
self = eazy.photoz.PhotoZ(param_file='cdfs.v1.6.1.zphot.param', translate_file=translate_file, zeropoint_file=None,
                         load_prior=True, load_products=False)
```

```python
#### Iterative zeropoint corrections

NITER = 3
NBIN = np.minimum(self.NOBJ//100, 180)

self.param.params['VERBOSITY'] = 1.
for iter in range(NITER):
    print('Iteration: ', iter)

    sn = self.fnu/self.efnu
    clip = (sn > 1).sum(axis=1) > 4 # Generally make this higher to ensure reasonable fits
    self.iterate_zp_templates(idx=self.idx[clip], update_templates=False,
                              update_zeropoints=True, iter=iter, n_proc=8,
                              save_templates=False, error_residuals=False,
                              NBIN=NBIN, get_spatial_offset=False)
```

```python
#### Now fit the whole catalog

# Turn off error corrections derived above
self.set_sys_err(positive=True)

# Full catalog
full = np.isfinite(self.cat['z_spec'])

# fit_parallel renamed to fit_catalog 14 May 2021
self.fit_catalog(self.idx[full], n_proc=8)
```

```python
# Show zspec-zphot comparison
fig = self.zphot_zspec()
```

```python
# Derived parameters (z params, RF colors, masses, SFR, etc.)
warnings.simplefilter('ignore', category=RuntimeWarning)
zout, hdu = self.standard_output(rf_pad_width=0.5, rf_max_err=2,
                                 prior=True, beta_prior=True)

# 'zout' also saved to [MAIN_OUTPUT_FILE].zout.fits
```
